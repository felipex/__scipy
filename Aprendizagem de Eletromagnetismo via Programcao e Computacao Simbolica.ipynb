```python
from sympy import *
x,y,z,f1,f2,f3 = symbols('x,y,z,f1,f2,f3')
X = [x,y,z]

def grad(f,n):
    a = []
    for x in X:
        a.append(f.diff(x,n))
    return a


def Div(Q):
    f1 = Q[0]
    f2 = Q[1]
    f3 = Q[2]
    
    return f1.diff(x) + f2.diff(y) + f3.diff(z)


def Laplacian(f):
    return Div(grad(f,1))


def curl(Q):
    f1 = Q[0]
    f2 = Q[1]
    f3 = Q[2]
    return [f3.diff(y) - f2.diff(z), f1.diff(z) - f3.diff(x), f2.diff(x) - f1.diff(y)]
```
