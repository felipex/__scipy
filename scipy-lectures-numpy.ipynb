## O que é NumPy e o que são Arrays NumPy?

### Arrays NumPy

O próprio Python já fornece arrays como elementos básicos da linguagem, mas o pacote NumPy traz uma nova implementação
que otimiza computacionalmente(experessão ruim) seu uso e já é projetado para o uso em computação científica.

Em Python puro um array é definido assim:

```python
L1 = range(10000)
L2 = [0,1,2,3]
```

Usando o NumPy fazemos assim:

```python
import numpy as np
a1 = np.arange(10000)
a2 = np.array([0,1,2,3])
```

Além da parte interna, a parte externa também tem diferenças entre os arrays. 

No exemplo abaixo vamos elevar todos os elementos tanto de L1 como de a1 para percebermos a diferença tanto no modo de chamar a operação como no seu desempenho.

```python
%timeit [i**2 for i in L1]
```

```python
%timeit a1**2
```

### Funções para criação de arrays

Na prática, dificilmente criamos um array elemento por elemento...

* Igualmente espaçados

```python
a = np.arange(10) # 0,1,2,..,n-1
a
```

```python
b = np.arange(1,9,2) # início, fim (exclusive), passo
b
```

* Ou por número de pontos

```python
c = np.linspace(0, 1, 6) # início, fim, número de pontos
c
```

```python
d = np.linspace(0, 1, 5, endpoint=False)
d
```

* Arrays comuns

```python
a = np.ones((3,3))
a
```

```python
b = np.zeros((3,3))
b
```

```python
c = np.eye(3)
c
```

```python
d = np.diag(np.array([1,2,3,4]))
d
```

* Números aleatórios  (Mersenne Twister PRNG):

```python
a = np.random.rand(4) # uniform in [0,1]
a
```

```python
b = np.random.randn(4) # gaussiana
b
```

```python
np.random.seed(1234) # configura a semente
```
